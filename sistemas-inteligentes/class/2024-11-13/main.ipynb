```python
def my_knn(
        features: list[list[int]],
        targets: list[int],
        input_features: list[int] | list[list[int]],
        k: int = 5 
    ) -> list[int]:

    if not isinstance(features, list) or not isinstance(targets, list) or not isinstance(input_features, list):
        raise ValueError("Os parÃ¢metros devem ser listas")
    
    if not isinstance(k, int):
        raise ValueError("k deve ser um inteiro")

    def euclidian_distance(
            x1:float, 
            x2:float
        ) -> float:
        if not isinstance(x1, (int, float)) or not isinstance(x2, (int, float)):
            raise ValueError("A distÃ¢ncia sÃ³ pode ser calculada entre nÃºmeros")
        
        return ((x1 - x2) ** 2)**0.5
    
    def features_distance(
            x1:list[float], 
            x2:list[float]
        ) -> float:

        if not isinstance(x1, list) or not isinstance(x2, list):
            raise ValueError("As features devem ser listas de nÃºmeros")
        
        distancies = [euclidian_distance(x1[i], x2[i]) for i in range(len(x1))]

        result = 0

        for dist in distancies:

            result += dist

        return result

    def calc_distance(
            features: list[list[float]],
            input_features: list[float],
            k: int
        ) -> tuple[list[float], list[int]]:

        distances = [float("inf") for _ in range(k)]

        classes = [-1 for _ in range(k)]

        for i in range(len(features)):

            distance = features_distance(features[i], input_features)

            for idx in range(len(distances)):
                
                if distance < distances[idx]:
                    
                    distances[idx] = distance
                    classes[idx] = targets[i]

                    break

        return distances, classes
    
    def predict(
            classes: list[int]
        ) -> int:

        greater_quantity = 0
        current_quantity = 0

        for i in range(len(classes)):
            for j in range(len(classes)):
                if classes[i] == classes[j]:
                    current_quantity += 1

            if current_quantity > greater_quantity:
                most_voted = classes[i]
                greater_quantity = current_quantity

            current_quantity = 0

        return most_voted

    def predicts(
            input_features: list[list[float]] | list[float]

        ) -> list[int]:

        result = []

        if isinstance(input_features[0], list):


            for input_feature in input_features:

                _, classes = calc_distance(features, input_feature, k)

                result.append(predict(classes))
        
        else:

            _, classes = calc_distance(features, input_features, k)

            result.append(predict(classes))

        return result
    
    return predicts(input_features)
```

```python
def my_accuracy(
        predicts: list[int],
        targets: list[int]
) -> float:
    
    if not isinstance(predicts, list) or not isinstance(targets, list):
        raise ValueError("Os parÃ¢metros devem ser listas")

    if len(predicts) != len(targets):
        raise ValueError("As listas devem ter o mesmo tamanho")
    
    corrects = 0

    total = len(predicts)

    for i in range(total):
        if predicts[i] == targets[i]:
            corrects += 1

    return corrects/total
```

```python
def test_predict_x_10(epoch: int = 10, ks:list[int] = [5]) -> dict[int, list[float]]:
    result = {}
    for k in ks:

        result[k] = []

        for _ in range(epoch):

            x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

            y_pred = my_knn(x_train, y_train, x_test, k)


            #print(f"\nK: {k}\nPredict:   {(y_pred)}\n expected: {(y_test)}")

            result[k].append(my_accuracy(y_pred, y_test))

    return result
```

```python
class MyKNN():
    def __init__(self, k: int):
        self.features = []
        self.targets = []
        self.k = k
        

    def predict(self, input_features: list[list[int]] | list[list[int]],

        ) -> list[int]:
        y_pred = my_knn(self.features, self.targets, input_features, k)
        return y_pred
    
    def fit(self, x_train, y_train):
        self.features = x_train
        self.targets = y_train
```

```python
def test(x: list[list[int]], y: list[int], classifies: list[RandomForestClassifier, SVC]):
    result_accuracy = np.zeros((3,5))
    result_precision = np.zeros((3,5))
    result_f1 = np.zeros((3,5))
    
    for i, (train_index, test_index) in enumerate(kfold.split(x)):
        x_train, x_test = np.array(x)[train_index], np.array(x)[test_index]
        y_train, y_test = np.array(y)[train_index], np.array(y)[test_index]

        x_train = x_train.tolist()
        y_train = y_train.tolist()
        x_test = x_test.tolist()

        for class_id, classify in enumerate(classifies):

            classify.fit(x_train, y_train)

            y_pred = classify.predict(x_test)

            result_accuracy[class_id][i] = round(accuracy_score(y_test, y_pred),2)
            result_precision[class_id][i] = round(precision_score(y_test, y_pred),2)
            result_f1[class_id][i] = f1_score(y_test, y_pred)

            
        

    return result_accuracy, result_precision, result_f1
```

```python
result_accuracy, result_precision, result_f1 = test(x, y, [MyKNN(5), RandomForestClassifier(), SVC()])
```
