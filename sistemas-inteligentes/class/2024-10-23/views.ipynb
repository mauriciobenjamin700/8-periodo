```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('regression_dataset.csv')
```

```python
df
```

```python
# Scatter plot do lucro em relaÃ§Ã£o aos perÃ­odos do dia
plt.figure(figsize=(10,6))

# Scatter para o perÃ­odo da manhÃ£
plt.scatter(df['morning'], df['profit'], color='green', label='Morning', alpha=0.6)

# Scatter para o perÃ­odo da tarde
plt.scatter(df['after_noon'], df['profit'], color='yellow', label='Afternoon', alpha=0.6)

# Scatter para o perÃ­odo da noite
plt.scatter(df['night'], df['profit'], color='blue', label='Night', alpha=0.6)

plt.title('Profit vs Time of Day')
plt.xlabel('Time of Day (0 = Not Present, 1 = Present)')
plt.ylabel('Profit')
plt.legend()
plt.grid(True)
plt.savefig('Profit vs Time of Day')
plt.show()
```

```python
# GrÃ¡fico de barras do lucro mÃ©dio por dia da semana
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
mean_profit_by_day = df[days].multiply(df['profit'], axis=0).mean()

plt.figure(figsize=(10,6))
mean_profit_by_day.plot(kind='bar', color='gray')
plt.title('Average Profit by Day of the Week')
plt.ylabel('Average Profit')
plt.xlabel('Day of the Week')
plt.grid(True)
plt.savefig('profit per week.png')
plt.show()
```

```python
# Boxplot do lucro em relaÃ§Ã£o aos meses do ano
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df_melted = df.melt(id_vars=['profit'], value_vars=months, var_name='Month', value_name='Presence')

plt.figure(figsize=(10,6))
sns.boxplot(x='Month', y='profit', data=df_melted)
plt.title('Boxplot of Profit by Month (All Data)')
plt.ylabel('Profit')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
